La comprensión de cómo los clientes perciben los servicios proporcionados por las instituciones financieras se ha convertido en una prioridad clave para las entidades bancarias. En este contexto, surge la necesidad de explorar la percepción de los clientes respecto al servicio ofrecido por el Banco Caja Social

Este proyecto se centra en el análisis de sentimiento en las redes sociales para examinar y cuantificar la percepción de los usuarios en relación con la precisión de los servicios bancarios ofrecido por el Banco Caja Social.
